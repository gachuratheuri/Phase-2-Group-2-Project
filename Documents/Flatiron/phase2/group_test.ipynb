```python
#connecting to database
conn = sqlite3.connect("sales.db")
#If sales.db doesn't exist it will be created automatically.
cur =conn.cursor()
```

Creating Table Manually (Using Cursor)

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS sales (
    id INTEGER PRIMARY KEY,
    product TEXT,
    category TEXT,
    region TEXT,
    sales_rep TEXT,
    quantity INTEGER,
    price REAL,
    sale_date TEXT
)
""")

conn.commit()
```

1. View Tables

```python
cur.execute("""SELECT name FROM sqlite_master WHERE type = 'table';""")
# Fetch the result and store it in table_names
table_names = cur.fetchall()
table_names
```
