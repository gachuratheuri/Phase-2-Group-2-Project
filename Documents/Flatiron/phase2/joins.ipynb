```python
# libraries
import pandas as pd
import sqlite3
```

```python
# This creates the database file if it doesn't exist
conn = sqlite3.connect("school.db")
cur = conn.cursor()
```

1.Creating students table

```python
create_table_query = """
CREATE TABLE IF NOT EXISTS Students (
    StudentID INT PRIMARY KEY,
    Name VARCHAR(50),
    Age INT,
    City VARCHAR(50)
);
"""

cur.execute(create_table_query)
conn.commit()
```

2.Creating Courses table

```python
create_table_query = """ 
CREATE TABLE IF NOT EXISTS Courses (
    CourseID INT PRIMARY KEY,
    CourseName VARCHAR(100),
    Credits INT
);
"""
cur.execute(create_table_query)
conn.commit()
```

3. Creating Enrollment table

```python
create_table_query = """ 
CREATE TABLE Enrollments (
    EnrollmentID INT PRIMARY KEY,
    StudentID INT,
    CourseID INT,
    EnrollmentDate DATE,
    FOREIGN KEY (StudentID) REFERENCES Students(StudentID),
    FOREIGN KEY (CourseID) REFERENCES Courses(CourseID)
);
"""
cur.execute(create_table_query)
conn.commit()
```

```python
#Inner Join of the three tables
q =""" 
SELECT s.Name, c.CourseName 
FROM Students s 
INNER JOIN Enrollments e
ON s.StudentID = e.StudentID
INNER JOIN Courses c
ON e.CourseID = c.CourseID

"""
pd.read_sql(q, conn)
```

```python
#Full outer Join students table and enrollments table
#Using UNION ALL
q = """  
SELECT s.StudentID, s.Name, e.CourseID
FROM Students s
LEFT JOIN Enrollments e
ON s.StudentID = e.StudentID
UNION  
SELECT s.StudentID, s.Name, e.CourseID
FROM Enrollments e
LEFT JOIN Students s
ON e.StudentID = s.StudentID;
"""
pd.read_sql(q, conn)
```

SUBQUERIES

```python
#SUBQUERIES INSIDE A QUERY
q= '''
   SELECT s.Name,
           (SELECT COUNT(*)
           FROM Enrollments e
           WHERE e.StudentID = s.StudentID)AS TotalCourses
           FROM Students s;
    '''                                        
```

```python
pd.read_sql(q,conn)
```

```python
#subquery in the From statement
#Subquery in the FROM Statement

q = '''
        SELECT Name, TotalCourses
        FROM (
                SELECT s.Name, COUNT(e.CourseID) AS TotalCourses
                FROM Students s 
                LEFT JOIN Enrollments e
                ON s.StudentID = e.StudentID
                GROUP BY s.Name
                )
        ORDER BY TotalCourses DESC;
    
     ''' 
pd.read_sql(q, conn)
```

```python
# subquery in the where statement
q = '''
        SELECT Name
        FROM Students
        WHERE StudentID IN (
                             SELECT StudentID
                             FROM Enrollments
                             WHERE CourseID >= 101)
    ;
     ''' 
pd.read_sql(q, conn)
```

```python
#finding students enrolled in Data analysis
q ='''
   SELECT Name
   FROM Students s
   INNER JOIN Enrollments e
     ON s.StudentID = e.StudentId
   INNER JOIN Courses c
     ON e.CourseID = c.CourseID
   WHERE c.CourseName="Data Analysis"
   ;
   '''
pd.read_sql(q, conn)
```

```python
q = '''
SELECT s.Name
FROM Students s
INNER JOIN Enrollments e
    ON s.StudentID = e.StudentID
INNER JOIN Courses c
    ON e.CourseID = c.CourseID
WHERE c.CourseName = "Data Analysis"
'''
pd.read_sql(q, conn)
```

```python
q = '''
SELECT Name
FROM Students
WHERE StudentID IN (
    SELECT StudentID
    FROM Enrollments
    WHERE CourseID = (
        SELECT CourseID
        FROM Courses
        WHERE CourseName = "Data Analysis"
    )
)
'''
pd.read_sql(q, conn)
```

```python
#view is a virtual Table
#it can be defined as a saved sql query stored in a database
conn.execute('''
CREATE VIEW studentcourseview As 
SELECT s.Name,c.CourseName,e.Enrollment
FROM Students s
INNER JOIN Enrollments e ON s.StudentID = e.StudentID
INNER JOIN Courses c ON e.CourseID = c.CourseID
''')
```
